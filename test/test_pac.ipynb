```python
# %load test_pac.py
from pathological import pac
import numpy as np

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt; plt.ion

spikes = pac.Spikes(50, 2)
times = spikes.times  # brevity

# Create biases
osc = pac.osc(times, 2.1, 12)
stim = pac.stim(times, 30, 1)
gain = osc * stim

# Simulate spiking
osc_sp = spikes.poisson(osc)
stim_sp = spikes.poisson(stim)
gain_sp = spikes.poisson(gain)

# Analysis
ts1, ns1 = pac.to_spiketimes(times, osc_sp)
plt.subplot(411)
plt.plot(ns1, ts1, 'o')

ts2, ns2 = pac.to_spiketimes(times, stim_sp)
plt.subplot(412)
plt.plot(ns2, ts2, 'o')

ts2, ns2 = pac.to_spiketimes(times, gain_sp)
plt.subplot(413)
plt.plot(ns2, ts2, 'o')

plt.subplot(414)
plt.plot(times, osc, label='osc')
plt.plot(times, stim, label='stim')
plt.plot(times, gain, label='gain')
plt.legend(loc='best')
```
