```python
import matplotlib.pyplot as plt
%matplotlib inline
from pykdf.kdf import load_kdf

import os, sys
import numpy as np
```

```python
csv_path = "/home/ejp/src/pacological/ipynb/"

fi = np.loadtxt(os.path.join(csv_path, "fi.csv"), delimiter=",")
g = np.loadtxt(os.path.join(csv_path, "g.csv"), delimiter=",")
print fi.shape, g.shape

I_fis = np.linspace(0, 50e-3, 100) * 1000
```

```python
k = 3
plt.plot(I_fis, fi[k, :], 'k')
plt.plot(I_fis, g[k, :], 'grey')
plt.plot(I_fis, g[k, :] * fi[k, :], 'blue')
```

```python

rt_e = []
rt_i = []
for k in range(4000):
    rt = np.trapz(g[k, :] * fi[k, :], I_fis)
    if k % 2 -- 0:
        rt_e.append(rt)
    else:
        rt_i.append(rt)
```

```python
plt.plot(rt_e, 'k')
plt.plot(rt_i, 'r')
```
